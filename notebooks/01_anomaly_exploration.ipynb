# Day 22 · Anomaly Detection in Microbiome Samples

```python
import sys; sys.path.insert(0,'..') 
from src.data.simulator import build_dataset
from src.models.detector import AnomalyDetector
X_df, y_s = build_dataset(n_normal=100, n_anomaly_per_type=10)
X_normal = X_df.values[(y_s=='normal').values]
det = AnomalyDetector(contamination=0.15)
det.fit(X_normal)
scores = det.score_samples(X_df.values)
print('Top anomalous:', X_df.index[scores['isolation_forest'].argsort()[::-1][:5]].tolist())
```
